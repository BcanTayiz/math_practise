```python
import seaborn as sns
import matplotlib.pyplot as plt
import math
import random
import sympy
from sympy import primepi
import numpy as np
```

```python
class PrimeSum:

    def __init__(self,n):
        self.n = n
        self.sum = 0
        self.primes = []

    def determine_sum(self):
        n = self.n
        prime_count = (n / math.log(n))

        y = n // 2

        x = prime_count
        val = (x * (x+1) / 2) # this should be the lower thereshold
        val = math.sqrt(y * val) * math.sqrt(n)

        if len(self.primes) > 0:
            self.primes.extend(list(sympy.primerange(self.primes[-1],n)))
        else:
            self.primes = list(sympy.primerange(0,n))
        #print(f"n value is {n}, x value is {x}")
        sum_of_primes = sum(self.primes)
        #print(f"prime sum is {sum_of_primes} and value is {val}")
        return val,sum_of_primes
      

    def plot_arrays(self,arr,arr2, title="Array Plot"):
        sns.set_theme(style="darkgrid")
        sns.lineplot(x=range(len(arr)), y=arr)
        sns.lineplot(x=range(len(arr2)), y=arr2)
        plt.title(title)
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()


```

```python
vals_1 = []
vals_2 = []
for i in range(10**5,10**5+100):

    primeSum = PrimeSum(i)

    val,sum_of_primes = primeSum.determine_sum()
    vals_1.append(val)
    vals_2.append(sum_of_primes)

primeSum.plot_arrays(vals_1,vals_2)
```
