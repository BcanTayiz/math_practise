```python
import seaborn as sns
import matplotlib.pyplot as plt
import math
import random
import sympy
from sympy import primepi
import numpy as np
```

```python
class PrimeSum:

    def __init__(self,n):
        self.n = n
        self.sum = 0
        self.primes = []

    def determine_sum(self):
        n = self.n
        prime_count = (n / math.log(n))

        y = n // 2

        x = prime_count
        val = (x * (x+1) / 2) # this should be the lower thereshold
        under_val = (math.sqrt(y* val) * math.sqrt(n+prime_count)) + x**((y+(prime_count)*math.pi)/y)
        upper_val = under_val + (val) //2 
        

        val = (under_val + upper_val) / 2

        if len(self.primes) > 0:
            self.primes.extend(list(sympy.primerange(self.primes[-1],n)))
        else:
            self.primes = list(sympy.primerange(0,n))
        #print(f"n value is {n}, x value is {x}")
        sum_of_primes = sum(self.primes)
        #print(f"prime sum is {sum_of_primes} and value is {val}")
        return val,sum_of_primes,under_val,upper_val
      

    def plot_arrays(self,arr,arr2,arr3,arr4 ,title="Array Plot"):
        sns.set_theme(style="darkgrid")
        sns.lineplot(x=range(len(arr)), y=arr )
        sns.lineplot(x=range(len(arr2)), y=arr2)
        sns.lineplot(x=range(len(arr3)), y=arr3 )
        sns.lineplot(x=range(len(arr4)), y=arr4)
        plt.title(title)
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()


```

```python
vals_1 = []
vals_2 = []
under_vals = []
upper_vals = []
for i in range(10,5000):

    primeSum = PrimeSum(i)

    val,sum_of_primes,under_val,upper_val= primeSum.determine_sum()
    vals_1.append(val)
    vals_2.append(sum_of_primes)
    under_vals.append(under_val)
    upper_vals.append(upper_val)

primeSum.plot_arrays(vals_1,vals_2,under_vals,upper_vals)
```
